```python
import pandas as pd
from IPython.display import display
import string
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'figure.max_open_warning': 0})
from decimal import Decimal

from orangecontrib.bioinformatics.geneset.__init__ import list_all
from orangecontrib.bioinformatics.utils.statistics import Hypergeometric

import helper_enrichment as enr
from helper import PATH_RESULTS
```

```python
# Reference genes and conversion of gene names to EID
genes = pd.read_csv(path_genes + 'mergedGenes_RPKUM.tsv', sep='\t', index_col=0)
ALL_GENE_NAMES= set(genes[(genes != 0).any(axis=1)].index)
NAME_EID=enr.name_genes_entrez(gene_names=ALL_GENE_NAMES, key_entrez=False)
ALL_GENE_NAMES_EID=enr.convert_EID(genes=ALL_GENE_NAMES, name_EID=NAME_EID)
```

### Milestones

```python
data_deseq=pd.read_table(path_milestones+'DE_combined.tsv',index_col=0)
data_impulse=pd.read_table(path_milestones+'DEacrossMainStages_AX4_summary_fdr0.001.tsv',index_col=0)
data=pd.concat([data_deseq,data_impulse],sort=True,axis=1)
```

```python
comparisons=[comparison.replace('_FDR_overall','') for comparison in data.columns[
    data.columns.str.contains('_FDR_overall')]]
max_fdr_de=0.01
min_abs_lfc_de=2
```

```python
%matplotlib agg
for comparison in comparisons:
    for lFC_filter in ['>= ','<= -']:
        query_names=list(data.query( comparison + '_log2FoldChange ' +lFC_filter+ str(min_abs_lfc_de) +
                          ' & ' + comparison + '_FDR_overall <= ' + str(max_fdr_de) +
                          ' & ' + comparison + ' == 1').index)
        comparison_title=comparison+' lFC '+lFC_filter+ str(
            min_abs_lfc_de)+' and FDR_overall <= ' + str(max_fdr_de)
        if(len(query_names)>0):
            enriched=enr.group_diff_enrichment(query_names=query_names,
                                               group=comparison_title,
                                               name_eid=NAME_EID,all_gene_names_eid=ALL_GENE_NAMES_EID, 
                                               gene_sets_ontology=GENE_SETS_ONTOLOGY,
                                               use_annotated_genes=True,
                                               min_overlap=MIN_OVERLAP,padj=PADJ,
                                               make_enrichment_bar=True,
                                               max_FE_bar=MAX_FE_BAR, min_FDR_bar=MIN_FDR_BAR,
                                               cmap_FDR_bar=CMAP_FDR_BAR,lFDR_base_bar=LFDR_BASE_BAR
                                              )
            display(enriched[1][0])
        else:
            print('***  '+comparison_title+' selected: 0\n')
plt.close('all')  
```

### Disaggregation

#### DE in Nichols, et al. (2020) data: media at 0.5-2hr vs all buffer samples

```python
query_data=pd.read_table(
    path_disaggregation+'mediaVSbufferAll_alternativegreater_FDRoptim0.01_DE_media0.5hr1hr2hr_ref_bufferAll.tsv',
    index_col=0)
filter_value='padj<=0.01 & log2FoldChange >= 2'
de_mb=set(query_data.query(filter_value).index)
```

```python
%matplotlib agg
figs=[]
enriched=enr.group_diff_enrichment(query_names=de_mb,
                                   group='media 0.5-2hr vs buffer all',
                                   name_eid=NAME_EID,all_gene_names_eid=ALL_GENE_NAMES_EID, 
                                   gene_sets_ontology=GENE_SETS_ONTOLOGY,
                                   use_annotated_genes=True,
                                   min_overlap=MIN_OVERLAP,padj=PADJ,
                                   make_enrichment_bar=True,
                                   max_FE_bar=MAX_FE_BAR, min_FDR_bar=MIN_FDR_BAR,
                                   cmap_FDR_bar=CMAP_FDR_BAR,lFDR_base_bar=LFDR_BASE_BAR
                                  )
figs.append(enriched[1])
display(enriched[1][0])
plt.close('all')    
```

```python
# Check overlap of disaggreagtion genes with genes obtained on Nichols, et al. (2020) data 
hypergeometric=Hypergeometric()
def nichols_overlap(gene_names):
    ref_overlap=ALL_GENE_NAMES & de_mb
    query_overlap = set(gene_names) & de_mb
    p=hypergeometric.p_value(len(query_overlap), len(ALL_GENE_NAMES), len(ref_overlap), len(gene_names))
    print("DE on data from Nichols, et al. (2020) (deMB): \ndeMB overlap with nonzero genes: %.1f%% and group: %.1f%%; p-val: %.2E\n" % (
            (len(ref_overlap) / len(ALL_GENE_NAMES)) * 100, (len(query_overlap)/len(gene_names)) * 100,Decimal(str(p))))
    
```

#### Upregulated between specific timepoints in tgrB1 and tgrB1C1, but not in AX4, tagB, or comH

```python
def de_genes(file, filter_str):
    return set(pd.read_table(file,index_col=0).query(filter_str).index)

%matplotlib agg
for times in [[4,6],[6,8],[8,12]]:
    filter_value='padj<=0.01 & log2FoldChange >= 1.32'
    de_AX4=de_genes(file=path_disaggregation+'AX4_confoundRep_FDRoptim0.01_DE_hr'+str(times[1])+
        '_ref_hr'+str(times[0])+'.tsv', filter_str=filter_value)
    de_tagB=de_genes(file=path_disaggregation+'comH_confoundRep_FDRoptim0.01_DE_hr'+str(times[1])+
        '_ref_hr'+str(times[0])+'.tsv', filter_str=filter_value)
    de_comH=de_genes(file=path_disaggregation+'tagB_confoundRep_FDRoptim0.01_DE_hr'+str(times[1])+
        '_ref_hr'+str(times[0])+'.tsv', filter_str=filter_value)
    de_tgrB1=de_genes(file=path_disaggregation+'tgrB1_confoundRep_FDRoptim0.01_DE_hr'+str(times[1])+
        '_ref_hr'+str(times[0])+'.tsv', filter_str=filter_value)
    de_tgrB1C1=de_genes(file=path_disaggregation+'tgrB1C1_confoundRep_FDRoptim0.01_DE_hr'+str(times[1])+
        '_ref_hr'+str(times[0])+'.tsv', filter_str=filter_value)
    query_names=(de_tgrB1 & de_tgrB1C1)-( de_AX4 |de_tagB | de_comH)
    enriched=enr.group_diff_enrichment(query_names=query_names,
                                       group=str(times[1])+'hr vs '+str(times[0])+'hr',
                                       name_eid=NAME_EID,all_gene_names_eid=ALL_GENE_NAMES_EID, 
                                       gene_sets_ontology=GENE_SETS_ONTOLOGY,
                                       use_annotated_genes=True,
                                       min_overlap=MIN_OVERLAP,padj=PADJ,
                                       make_enrichment_bar=True,
                                       max_FE_bar=MAX_FE_BAR, min_FDR_bar=MIN_FDR_BAR,
                                       cmap_FDR_bar=CMAP_FDR_BAR, lFDR_base_bar=LFDR_BASE_BAR
                                      )
    display(enriched[1][0])
    
    nichols_overlap(query_names)
    
plt.close('all')    
```
